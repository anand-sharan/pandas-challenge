```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)

# Rename headers and provide cleaned names
school_data_df = school_data_df.rename(columns={"school_name": "School Name", "type": "Type",
                                          "size": "Size", "budget": "Budget"})

# Display first five rows of school data frame to output
school_data_df.head()
```

```python
# Read Student Data File and store into Pandas DataFrames
student_data_df = pd.read_csv(student_data_to_load)

# Rename headers and provide cleaned names
student_data_df = student_data_df.rename(columns={"student_name": "Student Name", "gender": "Gender",
                                            "grade": "Grade","school_name": "School Name",
                                            "reading_score": "Reading Score", "math_score":"Math Score"})

# Display first five rows of student data frame to output
student_data_df.head()
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["School Name", "School Name"])

# Display first five rows of school and student data frame to output
school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Create a data frame with district data
district_school_data_complete_df = school_data_complete_df.loc[school_data_complete_df["Type"] == "District"]

# Display first five rows of district school and student data frame to output
district_school_data_complete_df.head()
```

```python
# Calculate the total number of schools (District)
total_number_of_district_schools = len(district_school_data_complete_df["School ID"].unique())

# Display to output
total_number_of_district_schools
```

```python
# Calculate the total number of students (District)
total_number_of_district_students = len(district_school_data_complete_df["Student ID"].unique())

# Display to output
total_number_of_district_students
```

```python
# Calculate the total budget (District)
total_district_budget = sum(district_school_data_complete_df["Budget"].unique())

# Display to output
total_district_budget
```

```python
# Calculate the average math score (District)
district_avg_math_score = district_school_data_complete_df["Math Score"].mean()

# Display to output
district_avg_math_score
```

```python
# Calculate the average reading score
district_avg_reading_score = district_school_data_complete_df["Reading Score"].mean()

# Display to output
district_avg_reading_score
```

```python
# Find the total number of students with a passing math score (70 or greater)
district_total_number_of_students_70_or_greater_math_score = district_school_data_complete_df[district_school_data_complete_df["Math Score"] >= 70].count()

# Calculate the percentage of students with a passing math score (70 or greater)
# Ensure div by zero doesn't occur
if total_number_of_district_students > 0:
    district_percentage_of_students_70_or_greater_in_math_score = 100 * district_total_number_of_students_70_or_greater_math_score["Student ID"] / total_number_of_district_students
else:
    district_percentage_of_students_70_or_greater_in_math_score = 0

# Display to output
district_percentage_of_students_70_or_greater_in_math_score
```

```python
# Find the total number of students with a passing read score (70 or greater)
district_total_number_of_students_70_or_greater_read_score = district_school_data_complete_df[district_school_data_complete_df["Reading Score"] >= 70].count()

# Calculate the percentage of students with a passing read score (70 or greater)
# Ensure div by zero doesn't occur
if total_number_of_district_students > 0:
    district_percentage_of_students_70_or_greater_in_read_score = 100 * district_total_number_of_students_70_or_greater_read_score["Student ID"] / total_number_of_district_students
else:
    district_percentage_of_students_70_or_greater_in_read_score = 0

# Display to output
district_percentage_of_students_70_or_greater_in_read_score
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
district_total_number_of_students_70_or_greater_overall_passing = district_school_data_complete_df[(district_school_data_complete_df["Math Score"] >= 70)
                                                                    & (district_school_data_complete_df["Reading Score"] >= 70)].count()

# Calculate the percentage of students with a passing math and read score (70 or greater)
# Ensure div by zero doesn't cause error
if total_number_of_district_students > 0:
    district_percentage_of_students_70_or_greater_in_overall_passing = 100 * district_total_number_of_students_70_or_greater_overall_passing["Student ID"] / total_number_of_district_students
else:
    district_percentage_of_students_70_or_greater_in_overall_passing = 0

# Display to output
district_percentage_of_students_70_or_greater_in_overall_passing
```

```python
# Create a dictionary with the calulated district summary results
district_summary_dct = {"Total Schools": total_number_of_district_schools,
                        "Total Students": total_number_of_district_students,
                        "Total Budget": total_district_budget,
                        "Average Math Score": district_avg_math_score,
                        "Average Reading Score": district_avg_reading_score,
                        "% Students Passing Math": district_percentage_of_students_70_or_greater_in_math_score,
                        "% Students Passing Reading": district_percentage_of_students_70_or_greater_in_read_score,
                        "% Overall Passing": district_percentage_of_students_70_or_greater_in_overall_passing
                       }

# Create a new data frame from list of dictionary
district_summary_df = pd.DataFrame([district_summary_dct])

# Display district summary result to output
district_summary_df
```

```python

# Format the values: 
# Total Students: Add comma as thousands separator
# Total Budget: Prefix dollar sign ($) and display upto two decimal precision
# Average Math Score and Average Reading Score: Display upto two decimal precision
# Students Passing Math, Reading and Overall Passing: display upto two decimal precision
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["% Students Passing Math"] = district_summary_df["% Students Passing Math"].map("{:.2f}".format)
district_summary_df["% Students Passing Reading"] = district_summary_df["% Students Passing Reading"].map("{:.2f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.2f}".format)

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create a Group by data set with School Name and Type and aggregrate summary results and use as_index=True argument

grouped_school = school_data_complete_df.groupby(
                        ["School Name"], as_index=True).agg({"Type": ["max"],
                                                             "Student ID": ["count"],
                                                             "Budget": ["unique"],
                                                             "Math Score": ["mean"],
                                                             "Reading Score": ["mean"]
                                                             }
                                                            )

# Assign column header names as per the school summary report
grouped_school.columns = ["School Type",
                          "Total Students", "Total School Budget",
                          "Average Math Score", "Average Reading Score"]

# Convert Total Budget to float to display as numeric values
grouped_school["Total School Budget"] = grouped_school["Total School Budget"].astype(float)

# Display to output
grouped_school
```

```python
# Create a School Summary data frame from the grouby object
school_summary_df = pd.DataFrame(grouped_school)

# Add new columns to the School Summary data frame
summary_new_cols = ["Per Student Budget",
                    "% Passing Math",
                    "% Passing Reading",
                    "% Overall Passing"]

# Assign deafault zero value to new columns
school_summary_df[summary_new_cols] = pd.DataFrame([[0] * len(summary_new_cols)], index=school_summary_df.index)

# Display to output
school_summary_df.head()
```

```python
# Calculate the Per Student Budget

for index, row in school_summary_df.iterrows():
    if row["Total Students"] > 0:
        school_summary_df.loc[index,"Per Student Budget"] = row["Total School Budget"] / row["Total Students"]

# Display to output
school_summary_df.head()
```

```python
# Create a data frame of complete school and students whose Math Score is greater than or equal to 70
school_70_or_greater_math_score_df = school_data_complete_df.loc[school_data_complete_df["Math Score"] >= 70]

# Create a Group by data set with School Name and aggregrate Total students and use as_index=True argument
grouped_school_70_or_greater_math_score = school_70_or_greater_math_score_df.groupby(
                        ["School Name"], as_index=True).agg({"Student ID": ["count"]})

# Assign column header names as per the school summary report
grouped_school_70_or_greater_math_score.columns = ["Total Students with Math score >= 70"]

# Create a Summary data frame from the grouby object of complete school and students whose Math Score is >= 70
school_70_or_greater_math_score_summary_df = pd.DataFrame(grouped_school_70_or_greater_math_score)

# Display to output
school_70_or_greater_math_score_summary_df.head()
```

```python
# Create a data frame of complete school and students whose Reading Score is greater than 70
school_70_or_greater_read_score_df = school_data_complete_df.loc[school_data_complete_df["Reading Score"] >= 70]

# Create a Group by data set with School Name and aggregrate Total students and use as_index=True argument
grouped_school_70_or_greater_read_score = school_70_or_greater_read_score_df.groupby(
                        ["School Name"], as_index=True).agg({"Student ID": ["count"]})

# Assign column header names as per the school summary report
grouped_school_70_or_greater_read_score.columns = ["Total Students with Reading score >= 70"]

# Create a Summary data frame from the grouby object of complete school and students whose
# Reading Score is greater than or equal to 70
school_70_or_greater_read_score_summary_df = pd.DataFrame(grouped_school_70_or_greater_read_score)

# Display to output
school_70_or_greater_read_score_summary_df.head()
```

```python
# Create a data frame of complete school and students whose Math Score and Reading Score
# both is greater than or equal to 70 by performing an inner join
school_70_or_greater_math_and_read_score_df = pd.merge(
                                        pd.DataFrame(
                                            school_data_complete_df.loc[school_data_complete_df["Math Score"] >= 70]),
                                        pd.DataFrame(
                                            school_data_complete_df.loc[school_data_complete_df["Reading Score"] >= 70]),
                                             how="inner", on=["School Name","Student ID"])

# Create groupby object to grouby student count by school name
grouped_school_70_or_greater_math_and_read_score = school_70_or_greater_math_and_read_score_df.groupby(
                        ["School Name"], as_index=True).agg({"Student ID": ["count"]})

# Assign column header names as per the school summary report
grouped_school_70_or_greater_math_and_read_score.columns = ["Total Students with Math and Reading score >= 70"]

# Create a Summary data frame from the grouby object of complete school and students whose Math Score and Reading Score
# is greater than or equal to 70
school_70_or_greater_math_and_read_score_summary_df = pd.DataFrame(grouped_school_70_or_greater_math_and_read_score)

# Display to output
school_70_or_greater_math_and_read_score_summary_df.head()
```

```python
# To the School Summary data frame, merge the Total Students with
# Math Score >= 70, Reading Score >= 70 and Math and Reading Score both >= 70 

school_summary_df = pd.merge(school_summary_df, school_70_or_greater_math_score_summary_df,
                             how="inner", on=["School Name"])
school_summary_df = pd.merge(school_summary_df, school_70_or_greater_read_score_summary_df,
                             how="inner", on=["School Name"])
school_summary_df = pd.merge(school_summary_df, school_70_or_greater_math_and_read_score_summary_df,
                             how="inner", on=["School Name"])

# Display to output
school_summary_df.head()
```

```python
# Calculate the % Passing Math, % Passing Reading, % Overall Reading

for index, row in school_summary_df.iterrows():
    if row["Total Students"] > 0:
        school_summary_df.loc[index,"% Passing Math"] = 100 * row["Total Students with Math score >= 70"] / row["Total Students"]
        school_summary_df.loc[index,"% Passing Reading"] = 100 * row["Total Students with Reading score >= 70"] / row["Total Students"]
        school_summary_df.loc[index,"% Overall Passing"] = 100 * row["Total Students with Math and Reading score >= 70"] / row["Total Students"]        

# Display to output
school_summary_df.head()

```

```python
# Format the School Summary data frame by retaining only the school summary required columns
school_summary_final_df = school_summary_df.iloc[:,0:10]

# Organize school_summary_final_df
school_summary_final_df = school_summary_final_df[["School Type",
                                                   "Total Students",
                                                   "Total School Budget",
                                                   "Per Student Budget",
                                                   "Average Math Score",
                                                   "Average Reading Score",
                                                   "% Passing Math",
                                                   "% Passing Reading",
                                                   "% Overall Passing"]]

# Format the values: 
# Total School Budget: Prefix dollar sign ($), display upto two decimal precision and add comma as thousands separator
# Per Student Budget: Prefix dollar sign ($), display upto two decimal precision and add comma as thousands separator

school_summary_final_df["Total School Budget"] = school_summary_final_df["Total School Budget"].map("${:,.2f}".format)
school_summary_final_df["Per Student Budget"] = school_summary_final_df["Per Student Budget"].map("${:,.2f}".format)

# Set data frame index name to None
school_summary_final_df.index.name = None

# display to output
school_summary_final_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Start with displaying the school summary dataframe
school_summary_df.head()
```

```python
# Create a data frame for Scores by School Spending 
scores_by_school_spending_df = pd.DataFrame(school_summary_df.loc[:,
                                                                  ["Per Student Budget",
                                                                   "Average Math Score",
                                                                   "Average Reading Score",
                                                                   "% Passing Math",
                                                                   "% Passing Reading",
                                                                   "% Overall Passing"
                                                                  ]]
                                          )
# Display output
scores_by_school_spending_df.head()
```

```python
# Create the bins in which Data will be held
per_student_average_budget_ranges_bins = [0, 584, 629, 644, 675]

# Create the names for the four bins
group_names_per_student_average_budget_ranges_bins = ["<$584", "$585-629", "$630-644", "$645-675"]
```

```python
# Cut the dataframe and assign group names
scores_by_school_spending_df["Spending Ranges (Per Student)"] = pd.cut(scores_by_school_spending_df["Per Student Budget"],
                                                                  per_student_average_budget_ranges_bins,
                                                                  labels=group_names_per_student_average_budget_ranges_bins,
                                                                  include_lowest=True)
# Display output
scores_by_school_spending_df.head()
```

```python
# Groupy the Spending Ranges (Per Student) column
grouped_scores_by_school_spending = scores_by_school_spending_df.groupby(
          "Spending Ranges (Per Student)", as_index=True).agg({"Average Math Score": ["mean"],
                                                                "Average Reading Score": ["mean"],
                                                                "% Passing Math": ["mean"],
                                                                "% Passing Reading": ["mean"],
                                                                "% Overall Passing": ["mean"]})

# Assign column header names as per the school summary report
grouped_scores_by_school_spending.columns = ["Average Math Score",
                                             "Average Reading Score",
                                             "% Passing Math",
                                             "% Passing Reading",
                                             "% Overall Passing"]

# Create a Summary data frame from the grouby object
scores_by_school_spending_summary_df = pd.DataFrame(grouped_scores_by_school_spending)

# Display to output
scores_by_school_spending_summary_df
```

```python
# Format the column so that it displays upto two decimal places
scores_by_school_spending_summary_df["Average Math Score"] = scores_by_school_spending_summary_df["Average Math Score"].map("{:,.2f}".format)
scores_by_school_spending_summary_df["Average Reading Score"] = scores_by_school_spending_summary_df["Average Reading Score"].map("{:,.2f}".format)
scores_by_school_spending_summary_df["% Passing Math"] = scores_by_school_spending_summary_df["% Passing Math"].map("{:,.2f}".format)
scores_by_school_spending_summary_df["% Passing Reading"] = scores_by_school_spending_summary_df["% Passing Reading"].map("{:,.2f}".format)
scores_by_school_spending_summary_df["% Overall Passing"] = scores_by_school_spending_summary_df["% Overall Passing"].map("{:,.2f}".format)

# Display to output
scores_by_school_spending_summary_df
```
